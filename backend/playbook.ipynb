```python
import matplotlib.pyplot as plt
import requests
```

```python
locations = {
    'Haidhausen': (48.136096, 11.608942),
    'Ostbahnhof': (48.127208, 11.607957),
    'Schwabing': (48.177723, 11.591010),
    'Laim': (48.146124, 11.502870),
    'Flughafen': (48.3538857382193, 11.788059296296359),
    'Haupbahnhof': (48.14203459014511, 11.558140657633798),
    'Stachus': (48.14011918838394, 11.566868086469782),
    'Dachau': (48.26061816855299, 11.469966571125877),
    'Lehel': (48.141499264319066, 11.584606215191291),
    'Trudering': (48.12245453794639, 11.662685591910769),
    'Freising': (48.38157801033135, 11.755000315191301)
}
```

```python
origin = (48.20086835400505, 11.627775696209682)
origin_lat, origin_lon = origin

plt.figure(figsize=(7,7))

# Plot stations
for name, (lat, lon) in locations.items():
    plt.scatter(lon, lat, marker = 's', s=80)
    # Add label slightly offset so it doesn't overlap the point
    plt.text(lon + 0.002, lat + 0.002, name, fontsize=9)

# Plot origin
plt.scatter(origin_lon, origin_lat, s=150, linewidth=2)
plt.text(origin_lon + 0.002, origin_lat + 0.002, "Origin", fontsize=10, fontweight="bold")

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Station Map")
plt.grid(True)
plt.show()
```
